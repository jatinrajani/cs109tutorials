```python
np.random.seed(sum(map(ord,"distributions")))
```

```python
#using distplot
x=np.random.normal(size=100)
sns.distplot(x)
```

```python
#rugplot which draws small vertical tick at each observation
sns.distplot(x,kde=False,rug=True)
```

```python
#changing the binsize in distplot
sns.distplot(x,bins=10,kde=False,rug=True)
```

```python
#density plots can be drawn using distplot by setting hist off
sns.distplot(x,hist=False,rug=True)
```

```python
x=np.random.normal(0,1,size=30)
bandwidth=1.05*x.std()*x.size**(-1/5.)
support =np.linspace(-4,4,200)
kernels=[]
for xi in x:
    kernel=stats.norm(xi,bandwidth).pdf(support)
    kernels.append(kernel)
    plt.plot(support,kernel,color="r")
    
sns.rugplot(x,color=".2",linewidth=3)

```

```python

sns.kdeplot(x,shade=True)
```

```python
#fitting parametric distributions
x=np.random.gamma(5,size=200)
sns.distplot(x,kde=False,fit=stats.gamma)
```
